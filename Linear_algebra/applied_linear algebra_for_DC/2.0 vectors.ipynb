# Упражнение 2.3
Создайте функцию Python, которая на входе будет принимать вектор, а на выходе – выдавать единичный вектор в том же направлении. Что происходит при вводе вектора нулей?

```python
v = np.array([2, 3])
```

```python
# define function
def createUnitVector(v):
    # get vector norm
    mu = np.linalg.norm(v)
    # return unit vector
    return v / mu
```

```python
a = plt.arrow(
    0, 0, v[0], v[1], head_width=0.3, width=0.1, color="k", length_includes_head=True
)
a1 = plt.arrow(
    0,
    0,
    createUnitVector(v)[0],
    createUnitVector(v)[1],
    head_width=0.3,
    width=0.1,
    color="c",
    length_includes_head=True,
)
plt.grid(linestyle="--", linewidth=0.5)
plt.axis("square")
plt.axis([-6, 6, -6, 6])

plt.show()
```

## Упражнение 2.4
Вы знаете, как создавать единичные векторы; что, если вы хотите создавать вектор любого произвольного модуля? Напишите функцию Python, которая на входе будет принимать вектор и желаемый модуль вектора и возвращать вектор в том же направлении, но с модулем, соответствующим второму вход- ному аргументу.

```python
v = np.array([2, 3])
mo = 6
```

```python
def mod(v, m):
    return v / np.linalg.norm(v) * m
```

```python
mod(v, mo)
```

```python
a1 = plt.arrow(
    0, 0, v[0], v[1], head_width=0.3, width=0.1, color="k", length_includes_head=True
)
a = plt.arrow(
    0,
    0,
    mod(v, mo)[0],
    mod(v, mo)[1],
    head_width=0.3,
    width=0.1,
    color="b",
    length_includes_head=True,
)
plt.grid(linestyle="--", linewidth=0.5)
plt.axis("square")
plt.axis([-6, 6, -6, 6])

plt.show()
```

## Упражнение 2.5
Напишите цикл for для транспонирования вектора-строки в вектор-столбец без использования встроенной функции или метода, такого как np.transpose() или v.T. Данное упражнение поможет вам создавать и индексировать векто- ры с наделенной ориентацией.

```python
v = np.array([1, 2, 3, 4, 5])
```

```python
def transpose(v):
    a = np.zeros(shape=(5, 1))
    for i in range(v.shape[1]):
        a[i, 0] = v[0, i]
    return a
```

```python
transpose(v)
```
